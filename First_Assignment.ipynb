```python
import pandas as pd
import requests

coin_ids = ['bitcoin', 'ethereum', 'tether']
all_coins_historical_data_list = []

for coin_id in coin_ids:
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=inr&days=365"

    response = requests.get(url)
    historical_data = response.json()

    if 'prices' in historical_data:
        df_coin = pd.DataFrame(historical_data['prices'], columns=['timestamp_ms', 'price_inr'])
        df_coin['date'] = pd.to_datetime(df_coin['timestamp_ms'], unit='ms').dt.date
        df_coin['coin_id'] = coin_id
        all_coins_historical_data_list.append(df_coin)
    else:
        print(f"No price data found for {coin_id}")

top_coins = pd.concat(all_coins_historical_data_list, ignore_index=True)

top_coins = top_coins[['coin_id', 'date', 'price_inr']]
display(top_coins.head())
display(top_coins.tail())
```

**1 cryptocurrencies**

```python
#1. Find the top 3 cryptocurrencies by market cap.

query = """
SELECT id, symbol, name, market_cap
FROM cryptocurrencies
ORDER BY market_cap DESC
LIMIT 3;
"""
pd.read_sql(query, conn)
```

```python
#2. List all coins where circulating supply exceeds 90% of total supply.

query = """
SELECT id, symbol, name, circulating_supply, total_supply
FROM cryptocurrencies
WHERE total_supply IS NOT NULL
  AND circulating_supply >= 0.9 * total_supply;
"""

pd.read_sql(query, conn)
```

```python
#3. Get coins that are within 10% of their all-time-high (ATH).

query = """
SELECT id, symbol, name, current_price, ath
FROM cryptocurrencies
WHERE current_price <= 0.1 * ath;
"""

pd.read_sql(query, conn)
```

```python
#4. Find the average market cap rank of coins with volume above $1B.

query = """
SELECT AVG(market_cap_rank) AS avg_market_cap_rank
FROM cryptocurrencies
WHERE total_volume > 1000000000;
"""

pd.read_sql(query, conn)
```

```python
#5. Get the most recently updated coin.

query = """
SELECT id, symbol, name, date
FROM cryptocurrencies
ORDER BY date DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

**2 crypto_prices (daily prices of top coins)**

```python
#1. Find the highest daily price of Bitcoin in the last 365 days.

query = """
SELECT MAX(price_inr) AS highest_price
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND date >= date('now', '-365 day');
"""
pd.read_sql(query, conn)
```

```python
#2. Calculate the average daily price of Ethereum in the past 1 year.

query = """
SELECT AVG(price_inr) AS avg_price
FROM crypto_prices
WHERE coin_id = 'ethereum'
  AND date >= (
      SELECT date(MAX(date), '-1 year')
      FROM crypto_prices
      WHERE coin_id = 'ethereum'
  );
"""
pd.read_sql(query, conn)
```

```python
#3. Show the daily price trend of Bitcoin in January 2025.

query = """
SELECT date, price_inr
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND date BETWEEN '2025-01-01' AND '2025=01-31'
ORDER BY date;
"""
pd.read_sql(query, conn)
```

```python
#4. Find the coin with the highest average price over 1 year.

query = """
SELECT coin_id, AVG(price_inr) AS avg_price
FROM crypto_prices
WHERE date >= date('now', '-1 year')
GROUP BY coin_id
ORDER BY avg_price DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

```python
#5. Get the % change in Bitcoins price between Sep 2024 and Sep 2025.

query = """
SELECT
  (
    (end_price - start_price) * 100.0 / start_price
  ) AS percentage_change
FROM (
  SELECT
    (SELECT price_inr
     FROM crypto_prices
     WHERE coin_id = 'bitcoin'
       AND date >= '2025-09-01'
     ORDER BY date ASC
     LIMIT 1) AS start_price,

    (SELECT price_inr
     FROM crypto_prices
     WHERE coin_id = 'bitcoin'
       AND date <= '2025-09-30'
     ORDER BY date DESC
     LIMIT 1) AS end_price
);
"""

pd.read_sql(query, conn)
```

```python
#4. Find the lowest price of oil in the last 10 years.

query = """
SELECT
  MIN(price_inr) AS lowest_price_last_10_years
FROM oil_prices
WHERE date >= (
  SELECT date(MAX(date), '-3years') FROM oil_prices
);
"""
pd.read_sql(query, conn)
```

```python
#5. Calculate the volatility of oil prices (max-min difference per year).

query = """
SELECT
  strftime('%Y', date) AS year,
  MAX(price_inr) - MIN(price_inr) AS yearly_volatility
FROM oil_prices
GROUP BY year
ORDER BY year;
"""
pd.read_sql(query, conn)
```

# ** Join Queries (Cross-Market Analysis)**

```python
#1. Compare Bitcoin vs Oil average price in 2025.

query = """
SELECT
  AVG(CASE WHEN cp.coin_id = 'bitcoin' THEN cp.price_inr END) AS btc_avg_price,
  AVG(op.price_inr) AS oil_avg_price
FROM crypto_prices cp
JOIN oil_prices op
  ON cp.date = op.date
WHERE cp.coin_id = 'bitcoin'
  AND cp.date BETWEEN '2025-01-01' AND '2025-12-31';
"""
pd.read_sql(query, conn)
```

```python
#2. Check if Bitcoin moves with S&P 500 (correlation idea).

query = """
SELECT
  cp.date,
  cp.price_inr AS btc_price,
  sp.close AS sp500_close
FROM crypto_prices cp
JOIN stock_prices sp
  ON cp.date = sp.date
WHERE cp.coin_id = 'bitcoin'
  AND sp.ticker = '^GSPC'
ORDER BY cp.date;
"""
pd.read_sql(query, conn)
```

```python
#3. Compare Ethereum and NASDAQ daily prices for 2025.

query = """
SELECT
  cp.date,
  cp.price_inr AS eth_price,
  sp.close AS nasdaq_close
FROM crypto_prices cp
JOIN stock_prices sp
  ON cp.date = sp.date
WHERE cp.coin_id = 'ethereum'
  AND sp.ticker = '^IXIC'
  AND cp.date BETWEEN '2025-01-01' AND '2025-12-31'
ORDER BY cp.date;
"""
pd.read_sql(query, conn)
```

```python
#4. Find days when oil price spiked and compare with Bitcoin price change.

query = """
WITH oil_changes AS (
  SELECT
    date,
    price_inr,
    LAG(price_inr) OVER (ORDER BY date) AS prev_price
  FROM oil_prices
)

SELECT
  oc.date,
  oc.price_inr,
  cp.price_inr AS btc_price
FROM oil_changes oc
JOIN crypto_prices cp
  ON oc.date = cp.date
WHERE cp.coin_id = 'bitcoin'
  AND ((oc.price_inr - oc.prev_price) * 100.0 / oc.prev_price) > 5;
"""
pd.read_sql(query, conn)
```

```python
#5. Compare top 3 coins daily price trend vs Nifty (^NSEI)

query = """
SELECT cp.date,
       cp.coin_id,
       cp.price_inr AS crypto_price,
       sp.close AS nifty_close
FROM crypto_prices cp
JOIN stock_prices sp
ON cp.date = sp.date
AND sp.ticker = '^NSEI'
WHERE cp.coin_id IN ('bitcoin','ethereum','tether')
ORDER BY cp.date, cp.coin_id;
"""
pd.read_sql(query, conn)
```

```python
#6. Compare stock prices (^GSPC) with crude oil prices on the same dates.

query = """
SELECT
    s.date,
    s.close AS sp500_close,
    o.price_inr AS oil_price
FROM stock_prices s
JOIN oil_prices o
    ON s.date = o.date
WHERE s.ticker = '^GSPC'
ORDER BY s.date;
"""
pd.read_sql(query, conn)
```

```python
#7. Correlate Bitcoin closing price with crude oil closing price (same date)

query = """
SELECT
    c.date,
    c.price_inr AS bitcoin_price,
    o.price_inr AS oil_price
FROM crypto_prices c
JOIN oil_prices o
    ON c.date = o.date
WHERE c.coin_id = 'bitcoin'
ORDER BY c.date;
"""
pd.read_sql(query, conn)
```

```python
#8. Compare NASDAQ (^IXIC) with Ethereum price trends

query = """
SELECT
    s.date,
    s.close AS nasdaq_close,
    c.price_inr AS ethereum_price
FROM stock_prices s
JOIN crypto_prices c
    ON s.date = c.date
WHERE s.ticker = '^IXIC'
AND c.coin_id = 'ethereum'
ORDER BY s.date;
"""
pd.read_sql(query, conn)
```

```python
#9. Join top 3 crypto coins with stock indices for 2025

query = """
SELECT
    c.date,
    c.coin_id,
    c.price_inr,
    s.ticker,
    s.close AS stock_close
FROM crypto_prices c
JOIN stock_prices s
    ON c.date = s.date
WHERE c.coin_id IN ('bitcoin', 'ethereum', 'tether')
AND strftime('%Y', c.date) = '2025'
ORDER BY c.date, c.coin_id;
"""
pd.read_sql(query, conn)
```

```python
#10. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison

query = """
SELECT
    s.date,
    s.ticker,
    s.close AS stock_close,
    o.price_inr AS oil_price,
    c.price_inr AS bitcoin_price
FROM stock_prices s
JOIN oil_prices o
    ON s.date = o.date
JOIN crypto_prices c
    ON s.date = c.date
WHERE c.coin_id = 'bitcoin'
ORDER BY s.date;
"""
pd.read_sql(query, conn)
```
